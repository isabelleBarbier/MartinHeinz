```python
def get_data(page=1):
    url = f"https://api.themoviedb.org/3/movie/popular?language=fr-FR&page={page}"
    response = requests.get(url, headers=headers)
    return response.json()['results']
```

# Étape 3: Import des Données

```python
# Boucle pour récupérer les données de plusieurs pages
results = []
for page in range(1, 21):
    result = get_data(page)
    results.extend(result)
```

```python
# Création d'un DataFrame à partir des résultats agrégés
import pandas as pd
df = pd.DataFrame(results)
df
```

```python
# Conversion de la colonne de date en type de données datetime
df['release_date'] = pd.to_datetime(df['release_date'])
```
